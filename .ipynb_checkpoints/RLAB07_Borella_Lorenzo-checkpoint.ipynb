### A) What is the probability distribution of y, the number of times the new method fails to detect the disease ?

The probability for the test to correctly detect the disease is represented by the parameter $q$. The complementary value, the probability to fail in detecting the disease, is $p=1-q$. 

We have no prior knowledge about these parameters before collecting the data, nontheless we know that the probability for the test to fail in detecting the presence of the disease $k$ times, over a set of length $n$, follows the Binomial distribution:
$$P(k|n,p)={n\choose k} p^k q^{n-k}$$

I choosed to plot the Binomial distribution for the specific case $n=100,p=0.3$ in order to show the behaviour of the curve.


```r
x <- seq(0,100,1)

options(repr.plot.width=8, repr.plot.height=5)
plot(x,dbinom(x,size=100,prob=0.3),type="l",col="darkred",xlab="k",ylab="Probability Density",main="Binomial Distribution",lwd=3)
legend("topright",legend=c("Size=100","Probability to fail=0.3"))
```

### B) What is the frequentist estimator of the failure probability of the new method ?

The test was performed on a set of $n=75$ patients and only for $y=6$ cases the new method failed to detect the disease.
The best estimation we can make of p, the probability for the new method to fail in detecting the disease, in a frequentist manner is exactely $p=\frac{y}{n}$ with a standard deviation of $\sigma=\sqrt{\frac{p(1-p)}{n}}$, since it's the only information we can derive from the data we collected.

```r
n <- 75 #number of patients known to have the disease

y <- 6 #number of times the new method fails to detect the disease

p0 <- 0.15 #standard testing method

p <- y/n #frequentist estimator
sigma <- sqrt(p*(1-p)/n)

cat(sprintf("Standard testing method, p0=%.2f.\n",p0))
cat(sprintf("New method, p=%.2f\n",p))
```

### C) Setup a bayesian computation of the posterior probability, assuming a beta distribution with mean value 0.15 and standard deviation 0.14.

I calculated the $\alpha$ and $\beta$ parameters for the prior distribution with the usual formulae, considering $\mu=0.15$ and $\sigma=0.14$ as suggested by the exercise:
$$\alpha_{prior}=\mu(\frac{\mu(1-\mu)}{\sigma^2}-1)$$
$$\beta_{prior}=(1-\mu)(\frac{\mu(1-\mu)}{\sigma^2}-1)$$

The prior distribution has indeed the shape of a Beta distribution with the parameters $\alpha_{prior}$ and $\beta_{prior}$. Since the collected data suggest that the new test has failed in recognising the disease $y$ times over a set of $n$ people, the posterior distribution, collecting the informations derived by the data, will have again the shape of a Beta distribution but this time with parameters:
$$\alpha_{posterior}=\alpha_{prior}+y$$
$$\beta_{posterior}=\beta_{prior}+n-y$$ 

I plotted the posterior distribution highlighting the mean value and the variance. These values represent the best estimate of the parameter p, the probability for the test to fail in recognising the disease, obtained in a Beyesian manner. I choosed to plot the curve only in the range $p \in [0,0.25]$ since elsewhere it is identically equal to zero.

```r
#prior distribution
mean <- 0.15
sd <- 0.14
alpha_prior <- mean*(mean*(1-mean)/(sd^2)-1)
beta_prior <- (1-mean)*(mean*(1-mean)/(sd^2)-1)
prior <- Vectorize(function(x){dbeta(x,shape1=alpha_prior,shape2=beta_prior)}) #default normalized

#posterior
alpha_posterior <- alpha_prior+y
beta_posterior <- beta_prior+n-y
posterior <- Vectorize(function(t){dbeta(t,shape1=alpha_posterior,shape2=beta_posterior)}) #default normalized

#mean
mean <- integrate(function(t){t*posterior(t)},lower=0,upper=1)$value

#variance
variance <- integrate(function(t){posterior(t)*(t-mean)^2},lower=0,upper=1)$value

low <- mean-sqrt(variance)
up <- mean+sqrt(variance)

#plot
p <- seq(0,0.25,0.001)
options(repr.plot.width=8, repr.plot.height=5)
plot(p,posterior(p),type="l",col="dodgerblue4",main="Posterior Distribution",ylab="Probability Density",lwd=3)
polygon(x=c(low,low,p[(p>=low)&(p<=up)],up),y=c(0,posterior(low),posterior(p[(p>=low)&(p<=up)]),0), col="deepskyblue3")
abline(v=c(low,mean,up),col="black")
legend("topright",legend=c(sprintf("Mean=%.2f",mean),sprintf("Standard Deviation=%.2f",sqrt(variance))))

cat(sprintf("BAYESIAN:\nNew Method: p=%.2f +- %.2f",mean,sqrt(variance)))
```

### D) Perform a test of hypothesis. Test the sample at a 5% level of significance in the Bayesian way.

I defined the hypotheses in this way:
* $H_0:p\geq p_0$. The probability $p$ for the new test to fail in detecting the disease is greater than the one of the old test. The new method is not better than the old one and the data samples are solely due to a random fluctuation.
* $H_1:p<p_0$. The probability $p$ of failing for the new test is smaller than the one of the old test. The new method is indeed better than the old one and the data samples reflect this behaviour.

I fixed the significance at $\alpha=0.05$ and I calculated the probability for the hypothesis $H_0$ to be true by integrating the posterior distribution in the designed range:
$$P(p\geq p_0)=\int_{p_0}^1 P(p|k,n) dp$$


Since this probability happens to be smaller than the fixed significance value, we can reject the null hypothesis and assert that the new method is actually more efficient than the old one in detecting the disease. 


```r
#level of significance
alpha <- 0.05

#probability that the null hypothesis is TRUE, P(p>=p0)
PH0 <- integrate(function(t){posterior(t)},lower=p0,upper=1)$value
if(PH0<alpha){
    cat(sprintf("Significance=%.2f\n",alpha))
    cat(sprintf("P(p>=p0)=%.4f\n",PH0))
    cat("We can reject the null hypothesis.")}

#plot
p <- seq(0,1,0.001)
plot(p,posterior(p),type="l",col="dodgerblue4",main="Posterior Distribution",ylab="Probability Density",lwd=3)
polygon(x=c(p0,p0,p[p>=p0]),y=c(0,posterior(p0),posterior(p[p>=p0])),col="deepskyblue3")
abline(v=p0,col="black")
```

Since the former is smaller than the latter, as we can see from the plot I printed below, we can reject the null hypothesis.

If the probability to obtain what we obtained, calculated from the distribution in which the null hypothesis is meant to be true, is too small compared to our significance level $\alpha$, we can assert with a certain credibility that the sampled value $y=6$ cannot be due only to random fluctuation of the sampling distribution; the data in this case support the hypothesis that the new detecting method is actually more efficient than the old one.

### B) Using a Beta(1,10) prior for p, calculate the posterior distribution.

I calculated as usual the posterior distribution using Bayes theorem. In the particular case in which a Beta function is used as prior distribution, with parameters $\alpha_{prior}=1$ and $\beta_{prior}=10$, the posterior can be derived simply by updating these parameters with the values obtained from the data: 
$$\alpha_{posterior}=\alpha_{prior}+y_0=12 \\ \beta_{posterior}=\beta_{prior}+n_0-y_0=115$$

where, as before, $n_0=116$ is the number of total samples taken and $y_0=11$ is the number of samples that presented a high level of bacter X.

### C) Find the bayesian estimator for p, the posterior mean and variance, and a 95% credible interval.

I calculated the mean and the variance of p applying the respective definitions on the Posterior probability distribution:
$$\mu=\int_0^1 p*P(p|y) dp \\ \sigma^2=\int_0^1 (p-\mu)^2 P(p|y)dp$$

I plotted the normal approximation of the posterior, maintaining the previously derived mean and variance values; in this way I could derive the 95% credibility estimation of the parameter p. Eventually I printed the estimator of p derived in the Bayesian approach.


```r
#mean
mean <- integrate(function(t){t*posterior0(t)},lower=0,upper=1)$value

#variance
variance <- integrate(function(t){posterior0(t)*(t-mean)^2},lower=0,upper=1)$value

#norm approximation
norm_posterior <- Vectorize(function(p){dnorm(p,mean=mean,sd=sqrt(variance))})

#upper bound
upper_bayes <- qnorm(0.975,mean=mean,sd=sqrt(variance))

#lower bound
lower_bayes <- qnorm(0.025,mean=mean,sd=sqrt(variance))

#credibility interval
credibility <- integrate(function(t){dnorm(t,mean=mean,sd=sqrt(variance))},lower=lower_bayes,upper=upper_bayes)$value

#print the results
cat("BAYESIAN\n")
cat(sprintf("p=%.2f +- %.2f\n",mean,sqrt(variance)))
cat("Credibility interval:",credibility)

#plot
p <- seq(0,0.25,0.001)
plot(p,norm_posterior(p),type="l",col="darkred",main="Normal Approximation of Posterior",lwd=3,ylab="Probability Density")
legend("topright",legend=c(sprintf("Mean=%.2f",mean),sprintf("Standard Deviation=%.2f",sqrt(variance))))
polygon(x=c(lower_bayes,lower_bayes,p[(p>=lower_bayes)&(p<=upper_bayes)],upper_bayes),y=c(0,norm_posterior(lower_bayes),norm_posterior(p[(p>=lower_bayes)&(p<=upper_bayes)]),0),col="red")
abline(v=c(lower_bayes,mean,upper_bayes),col="black")
                                        
```

These values overlap within their standard deviations, but are indeed slightly different due to the different approaches followed to derive the posterior distribution. If we update our knowledge we are taking into consideration all the data collected, while if we consider only the second measurement we obtain a Posterior Distribution based on less informations.


```r
#BETA PRIOR

#norm approximation
norm_posterior <- Vectorize(function(p){dnorm(p,mean=mean_beta,sd=sqrt(variance_beta))})

#upper bound
upper_bayes <- qnorm(0.975,mean=mean_beta,sd=sqrt(variance_beta))

#lower bound
lower_bayes <- qnorm(0.025,mean=mean_beta,sd=sqrt(variance_beta))

#credibility interval
credibility <- integrate(function(t){dnorm(t,mean=mean_beta,sd=sqrt(variance_beta))},lower=lower_bayes,upper=upper_bayes)$value

#print the results
cat("BETA PRIOR:\n")
cat(sprintf("p=%.3f +- %.3f\n",mean_beta,sqrt(variance_beta)))
cat(sprintf("Credibility interval:%.2f\n",credibility))

#plot
p <- seq(0,0.25,0.001)
plot(p,norm_posterior(p),type="l",col="darkred",main="Beta Prior:Normal Approximation of Posterior",lwd=3,ylab="Probability Density")
legend("topright",legend=c(sprintf("Mean=%.3f",mean_beta),sprintf("Standard Deviation=%.3f",sqrt(variance_beta))))
polygon(x=c(lower_bayes,lower_bayes,p[(p>=lower_bayes)&(p<=upper_bayes)],upper_bayes),y=c(0,norm_posterior(lower_bayes),norm_posterior(p[(p>=lower_bayes)&(p<=upper_bayes)]),0),col="red")
abline(v=c(lower_bayes,mean_beta,upper_bayes),col="black")


#UPDATED PRIOR

#norm approximation
norm_posterior <- Vectorize(function(p){dnorm(p,mean=mean,sd=sqrt(variance))})

#upper bound
upper_bayes <- qnorm(0.975,mean=mean,sd=sqrt(variance))

#lower bound
lower_bayes <- qnorm(0.025,mean=mean,sd=sqrt(variance))

#credibility interval
credibility <- integrate(function(t){dnorm(t,mean=mean,sd=sqrt(variance))},lower=lower_bayes,upper=upper_bayes)$value

#print the results
cat("UPDATED PRIOR:\n")
cat(sprintf("p=%.3f +- %.3f\n",mean,sqrt(variance)))
cat("Credibility interval:",credibility)

#plot
p <- seq(0,0.25,0.001)
plot(p,norm_posterior(p),type="l",col="darkred",main="Updated Prior:Normal Approximation of Posterior",lwd=3,ylab="Probability Density")
legend("topright",legend=c(sprintf("Mean=%.3f",mean),sprintf("Standard Deviation=%.3f",sqrt(variance))))
polygon(x=c(lower_bayes,lower_bayes,p[(p>=lower_bayes)&(p<=upper_bayes)],upper_bayes),y=c(0,norm_posterior(lower_bayes),norm_posterior(p[(p>=lower_bayes)&(p<=upper_bayes)]),0),col="red")
abline(v=c(lower_bayes,mean,upper_bayes),col="black")

```

### H) Test the hypotesis $H_0: p=0.1$ versus $H_1:p\neq0.1$.

I performed the HT in the bayesian way considering the posterior derived by updating the prior distribution. Even with the updated informations, we still cannot reject the null hypothesis.

I plotted the posterior distribution and highlighted the rejection region ($\alpha=0.05$). Since the value $p=0.1$ lies in the acceptance region, we are not allowed to reject $H_0$.

By slightly modifying our significance level we could easily reject the hypothesis $H_0$, since the value $p=0.1$ is very near to the rejection region. This approach anyway would lead us to a decrease in the significance of our affirmation.

```r
#BAYESIAN APPROACH
if((0.1<=upper_bayes)&(0.1>=lower_bayes)){cat("BAYESIAN: We cannot reject the null hypothesis.\n")}

#plot
p <- seq(0,0.25,0.001)
plot(p,posterior1(p),type="l",col="darkred",main="BAYESIAN: Posterior Distribution with Rejection Region",lwd=3,ylab="Probability Density")
polygon(x=c(upper_bayes,upper_bayes,p[p>=upper_bayes],0.25),y=c(0,posterior1(upper_bayes),posterior1(p[p>=upper_bayes]),0),col="red")
polygon(x=c(0,p[p<=lower_bayes],lower_bayes,lower_bayes),y=c(0,posterior1(p[p<=lower_bayes]),posterior1(lower_bayes),0),col="red")
abline(v=c(lower_bayes,upper_bayes),col="black")
abline(v=0.1,col="red")
legend("topright",legend=c("Rejection bounds","p=0.1"),col=c("black","red"),lty=1:1)
```

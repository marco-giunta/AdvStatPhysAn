## Esercizio 1 - Blood test 

Un test in grado di stabilire la presenza di una malattia dall'analisi del sangue tende a fallire la predizione nel 15% dei pazienti effettivamente malati.
Durante uno screening vengono sottoposti al test $n=75$ persone che sanno già di avere la malattia:
- qual è la distribuzione di probabilità di y, dove y è il numero di volte che il nuovo metodo di screening fallisce nel verificare la malattia
- sui $n=75$ pazienti il metodo fallisce la predizione $y=6$ volte, qual è l'indicatore frequentista della probabilità di fallimento
- calcolare tramite inferenza baiesiana il posterior riguardo la probabilità di fallimento del test assumendo un beta prior $\mathbf{Beta}(\mu=0.15,\sigma=0.14)$
- effetture un test delle ipotesi assumendo che se la probabilità che il nuovo metodo ha di fallire la verifica della malattia in pazienti malati è maggiore o uguale del 15% allora non è migliore del primo considerato (5% level of significance).
- ripetere lo stesso test ma attraverso un approccio frequentista

Sia $y$ il numero di pazienti malati che ricevono un risultato negativo da uno screening, consideriamo questa come la nostra random variable che segue una distribuzione di probabilità binomiale. E' possibile pertanto rappresentare il problema tramite una likelihood uguale a 

$$ P(y|\theta) = \binom{n}{y} \theta^y(1-\theta)^{n-y} \ \ \ \ \text{con} \ n\ \text{uguale al numero di trials}$$

Il parametro su cui vogliamo agire con un processo di inferenza $\theta$  è il corrispondente del "bias" durante il lancio di una moneta. In questo caso è la probabilità che una persona con la malattia risulti negativa al test a cui è stata sottoposta. Essendo $y$ la nostra random variable la likelihood è il prodotto di probabilità di eventi indipendenti e mutuamente esclusivi.

Scegliamo un beta prior

$$ P(\theta) \sim \mathrm{Beta}(\mu = 0.15,\sigma = 0.14)$$

Il teorema di Bayes allora ci dice che 

$$ P(\theta|y) = \frac{P(y|\theta) \cdot P(\theta)}{P(y)} $$

Il denominatore rappresenta la probabilità di osservare i dati ovvero una media di ciascuna probabilità condizionata ad un modello pesata sul modello considerato, ovvero al variare del parametro $\theta$ che descrive il modello generativo. 
Osserviamo che in un caso come il nostro, determinare il termine di "evidence" (cioè il fattore di normalizzazione del posterior) non sia utile in quanto tale fattore è indipendente dal parametro su cui stiamo facendo inferenza.
Inoltre essendo la funzione $P(\theta)$ un "conjugate prior" alla likelihood binomiale il posterior che emerge dal processo di inferenza risulta una distribuzione 

$$ \mathrm{Beta}(\alpha', \beta') \ \ \ \text{con} \ \alpha' = \alpha + y, \beta' = \beta + n - y$$

Bisogna fare attenzione alla forma della distribuzione beta di cui conosciamo il valore atteso e la deviazione standard. Reparametrizziamo la distribuzione beta passando da $\alpha,\beta$ a $\mu,\text{var}$ 

$$ \alpha = \mu \left(\frac{\mu(1-\mu)}{\text{var}}-1\right), \text{ se } \text{var}< \mu(1-\mu)$$
$$ \beta = (1 - \mu)\left(\frac{\mu(1-\mu)}{\text{var}}-1\right), \text{ se }\text{var}< \mu(1-\mu) $$

```r
# verifichiamo che la reparametrizzazione della distribuzione beta sia possibile 
print(paste("var =",0.14^2))
print(paste("mu*(1-mu) =", (0.15*(1-0.15))))
print(paste("var < mu*(1-mu) =", 0.14^2 < 0.15*(1-0.15)))
```

```r
# Rappresentiamo adesso nel caso baiesiano la posterior distribution se n=75 e y=6 

n <-75
y <- 6

mu <- 0.15
std <- 0.14

alpha <- mu*(mu*(1-mu)/std^2-1)
beta <- (1-mu)*mu*(mu*(1-mu)/std^2-1)

theta <- seq(0,1, by=0.01) 
theta.prior <- c(seq(0,0.01, by=0.0001),seq(0.01,0.1,by=0.01),seq(0.1,0.9, by=0.1),seq(0.9,1,by=0.005))
# e' necessario costruire un altro vettore più fitto del precedente per rappresentare il prior in quanto altrimenti non vi sarebbero abbastanza
# punti 

binomial.likelihood <- function(theta,y=6,n=75) dbinom(x = y, size = n, prob = theta)
beta.prior <- function(theta) dbeta(theta,shape1=alpha,shape2=beta)
beta.posterior <- function(theta) dbeta(theta,shape1=alpha+y,shape2=beta+n-y)
map <- function(theta,posterior) theta[posterior==max(posterior)]


plot(theta.prior, beta.prior(theta.prior), col="dodgerblue", main="Beta prior and posterior", 
     xlab=expression(theta), ylab="P(theta|y)", type="l", lwd=3, ylim=c(0,14),xlim=c(0,1))
lines(theta, beta.posterior(theta), col="darkblue", xlab=expression(theta), type="l", lwd=3)
abline(v = map(theta, beta.posterior(theta)), col="blue", lwd=3)
legend("topright", legend = c("Beta prior", "Beta posterior", sprintf("Posterior mode=%.2f",map(theta, beta.posterior(theta)))) , 
       col=c("dodgerblue","darkblue","blue"), 
       fill=c("dodgerblue","darkblue","blue"))
```

Nel caso in cui invece volessimo trattare il problema con un approccio frequentista allora un "unbiased estimator" è 

$$ p_f = \frac{y}{n} \\ \mathrm{E}(p_f) = \frac{\mathrm{E}(y)}{n} = \frac{np}{n} = p $$

```r
pf <- y/n
plot(theta,binomial.likelihood(theta), type="l", col="green", lwd=3, xlab=expression(theta), ylab="P(y|theta)", 
     main="Likelihood frequentist estimator")
abline(v = pf, col="darkgreen", lwd=3)
legend("topright",legend=c("Likelihood",sprintf("MLE=%.2f",pf)), col=c("green","darkgreen"), fill=c("green","darkgreen"))
```

### Hypothesis testing 

La procedura di hypothesis testing è utile per verificare se una determinata condizione sul problema è determinata dallo stesso oppure è semplicemente dovuta al caso. In questa situazione consideriamo la possibilità che il nuovo test per diagnosticare la malattia sia migliore del precedente. Diremo dunque che esso si comporta meglio di quello vecchio quando la probabilità di sbagliare l'esito del test su pazienti con la malattia è minore del 15%, valore che aveva il primo test.

Consideriamo pertanto due ipotesi:
- la null hypothesis: $H_0: p \geq p_0$. Il nuovo test è peggiore del precedente ed è probabile che i dati misurati siano stati ottenuti a causa di fluttuazioni statistiche
- $H_1: p < p_0$. Il test si comporta meglio di quello iniziale


**Approccio Bayesiano** 

Innanzitutto fissiamo una "significance" $\alpha=0.05$ ovvero il rate entro il quale non rigettiamo la null hypothesis in quanto la probabilità che essa sia vera ha come soglia limite inferiore $\alpha$ . Essa infatti equivale alla probabilità di escludere $H_0$ nonostante sia valida.
Calcoliamo ora la posterior probability che la null hypothesis regga

$$P(H_0:p\geq p_0|y) = \int_{p_0}^{1} P(\theta|y) \mathrm{d}\theta$$

In questo integrale è necessario considerare i dati che sono stati osservati dal nuovo test ovvero che la random variable assuma il valore $y=6$ su $n=75$ ovvero bisogna usare il posterior calcolato tramite l'inferenza baiesiana.
Se il risultato è maggiore di $\alpha$ allora l'ipotesi nulla è accettata altrimenti è possibile rigettarla e quindi affermare che il nuovo test si comporti meglio del precedente in base alle osservazioni effettuate.

```r
significance <- 0.05
p0 <- 0.15
hyp.testing <- integrate(beta.posterior, p0,1)$value
sprintf("La probabilità che il nuovo risultato sia migliore del primo sia un effetto randomico è del %.2f %%",hyp.testing*100 )
if(hyp.testing < alpha) sprintf("Poichè %.3f è minore di %.3f il nuovo test si comporta meglio del precedente", hyp.testing, significance)

plot(theta, beta.posterior(theta), col="dodgerblue", main="Beta posterior", 
     xlab=expression(theta), ylab="P(theta|y)", type="l", lwd=3, ylim=c(0,14),xlim=c(0,0.4))
polygon(x=c(p0,p0,theta[theta>=p0]),y=c(0,beta.posterior(p0),beta.posterior(theta[theta>=p0])),col="deepskyblue3")
abline(v=p0,col="black")
legend("topright", legend=c(sprintf("Probabilità che H0 sia dovuta ad un effetto casuale = %.2f %%",hyp.testing*100)),
      col="deepskyblue3", fill="deepskyblue3")
```

**Approccio frequentista**

Al fine di costruire la statistica per testare l'ipotesi nulla considero la likelihood binomiale funzione della random variable $y$ con $\theta = 0.15$ ovvero utilizzo la verosimiglianza che descrive il processo della null hypothesis $H_0$. Se dovessimo trovare che la probabilità $P(y=6|\theta=0.15)$ è minore di $\alpha=0.05$ allora $H_0$ può essere rigettata

```r
x <- seq(1,30,1)
H0 <- Vectorize(function(y){dbinom(y,size=n,prob=p0)})
pvalue <- Vectorize(function(t){sum(dbinom(x=seq(0,100,1),size=n,prob=p0)[1:t])})

sprintf("La probabilità che il nuovo risultato sia migliore del primo sia un effetto randomico è del %.2f %%",pvalue(y)*100 )
if(hyp.testing < alpha) sprintf("Poichè %.3f è minore di %.3f il nuovo test si comporta meglio del precedente", pvalue(y), significance)

plot(x, H0(x),type="l", lwd=3, col="dodgerblue", xlab="y",ylab="P(y|theta=0.6)", main="Binomial likelihood")
polygon(x=c(0,x[x<=y],y,y),y=c(0,H0(x[x<=y]),H0(y),0),col="deepskyblue3")
abline(v=y,col="black")
legend("topright", legend=c(sprintf("Probabilità che H0 sia dovuta ad un effetto casuale = %.2f %%",pvalue(y)*100)),
      col="deepskyblue3", fill="deepskyblue3")
```

## Esercizio 2 - Poisson distribution likelihood

Il numero di cavalieri dell'armata prussiana morti durante un intervallo di tempo ($\Delta t$) a causa dei calci dei loro stessi cavalli segue una distribuzione di Poisson di parametro $\mu$. 

Sia $y$ la random variable corrispondente al numero di eventi misurati nell'intervallo di tempo $\Delta t$;  
La distribuzione di probabilità di Poisson è una distribuzione discreta la quale rappresenta come è distribuito il numero di eventi che accadono in intervalli di tempo successivi $\Delta t$.
Pertanto, nel processo di inferenza Baiesiana, identifichiamo la likelihood come:
$$P(\{y_j\} |\mu) = \prod_{j=1}^n f(y_j|\mu) \propto \mu^{\sum y_j} \cdot e^{-n \mu} $$

dove la funzione nel r.h.s corrisponde a 
$$f(y_j|\mu) = \frac{\mu^{y_j} \cdot e^{-\mu}}{y!} $$

E' possibile osservare come la funzione likelihood altro non sia che una distribuzione Gamma con parametri:

$$\mathbf{Gamma}(y|\alpha,\lambda) = k y^{\alpha -1}e^{-\lambda y}; \ \ \ \ \ \ \text{con} \ \ k = \frac{\lambda^\alpha}{\Gamma(\alpha)}$$

che confrontato con il nostro caso diventa: $\alpha = \sum y_j + 1$ e $\lambda = n$

Inoltre la distribuzione Gamma ha moda $\frac{\alpha - 1}{\lambda}$

### Uniform prior
Supponiamo inizialmente di non conoscere nulla riguardo la distribuzione del numero di eventi misurati nell'intervallo di dieci secondi. 
Necessariamente il prior sarà un prior uniforme, detto improprio in quanto se non viene limitato l'intervallo di definizione della random variable risulta non normalizzato. 
$$ g(\mu) = 1 \ \ \ \ \ \text{per} \ \ \mu > 0$$

Il posterior pertanto diventa

$$P(\mu|\{y_j\}) \propto P(\{y_j\} |\mu) \cdot g(\mu)  \propto \mu^{\sum y_j}e^{-n \mu} $$

Ovvero la distribuzione Gamma non cambia i propri parametri iniziali 

Prima di costruire l'apparato per risolvere il problema osserviamo come sono costituiti i dati che ci vengono forniti:

i set di osservazioni sono due e comprendono le frequenze del numero di morti in un dato intervallo di tempo. Come già visto in un esercizio precedente, utilizzare due osservazioni consecutivamente oppure tutte in una volta per aggiornare il posterior restituisce lo stesso risultato nella condizione in cui tali osservazioni sono di eventi indpendente, come in questo caso grazie alla proprietà della distribuzione di Poisson.
Pertanto assegno al vettore `deaths` una serie di `rep()` dei valori $y=0,1,2,3,4$ tanti quanti sono $n1+n2$ che si leggono in tabella.

```r
deaths <- c(rep(0,length=253),rep(1,length=156),rep(2,length=54),rep(3,length=14),rep(4,length=3))
x <- seq(0,4, length=1000)
hist(deaths,col="dodgerblue", freq=FALSE)
```

```r
uniform.prior <- function(x) return(rep(1,length = length(x)))
uniform.posterior <- function(data) return(c((sum(data) + 1), length(data)))
# la funzione uniform.posterior restituisce unicamente i parametri alpha e lambda della distribuzione gamma dopo l'update 
# scrivo una funzione anche per calcolare la moda per il plot.  
compute.mode.gamma <- function(alpha,lambda) return((alpha - 1)/lambda)

plot.gamma.posterior <- function(x,data,prior,posterior,compute.mode){
    alpha <- posterior(data)[1]
    lambda <- posterior(data)[2]
    plot(x,dgamma(x, shape=alpha, rate=lambda), type = "l", lwd = 3, col = "dodgerblue", xlab=expression(mu),
         ylab="P(mu|x)", main="Posterior",
         ylim=c(0,11))
    lines(x, prior(x), col = "green")
    mode <- compute.mode(alpha,lambda)
    abline(v = mode, col="red")
    legend("topright", legend = c("Prior","Posterior",sprintf("Mode=%.2f",mode)), col=c("green","dodgerblue","red"), 
           fill=c("green","dodgerblue","red"))
} 
```

```r
plot.gamma.posterior(x,deaths,uniform.prior,uniform.posterior,compute.mode.gamma)
sprintf("La distribuzione Gamma ha parametri alpha=%.2f e lambda=%.2f",uniform.posterior(deaths)[1], uniform.posterior(deaths)[2])
```

### Jeffrey's prior 
Analogamente al prior uniforme, il suo integrale sullo spazio dei parametri $\mu$ è infinito. Esso è inoltre invariante per ogni trasformazione continua di $\mu$ e la sua forma è:

$$ g(\mu) \propto \frac{1}{\sqrt(\mu)}$$

combinando con la likelihood vista prima otteniamo 

$$ P(\mu|\{x_j\}) \propto P(\{x_j\} |\mu) \cdot g(\mu)  \propto \mu^{\sum x_j - 1/2}e^{-n \mu}$$

che è di nuovo una distribuzione Gamma con $\alpha=\sum x_j + 1/2$ e $\lambda=n$

```r
# analogamente a quanto visto prima è necessario costruire le tre funzioni argomento della funzione plot.posterior cioè prior, posterior e 
# compute.mode (che in questo caso è la stessa essendo la distribuzione finale grazie ai conjugate prior sempre una distribuzione gamma)

jeffrey.prior <- function(x) return(1/sqrt(x))
jeffrey.posterior <- function(data) return(c((sum(data) + 1/2), length(data))) 
compute.mode.gamma <- function(alpha,lambda) return((alpha - 1)/lambda)
```

```r
plot.gamma.posterior(x,deaths,jeffrey.prior,jeffrey.posterior,compute.mode.gamma)
sprintf("La distribuzione Gamma ha parametri alpha=%.2f e lambda=%.2f",jeffrey.posterior(deaths)[1], jeffrey.posterior(deaths)[2])
```

Valutiamo per entrambi la media, varianza e moda del posterior sia analiticamente che numericamente.

Analiticamente, per una distribuzione Gamma, sappiamo che
$$\text{la media vale} \ \  \frac{\alpha}{\lambda}, \ \ \text{la varianza} \ \ \frac{\alpha}{\lambda^2}$$

```r
# mentre le funzioni posterior definite nella cella precedente sono state utili per evitare di calcolare la distribuzione gamma 
# ogni volta che ce ne fosse bisogno, le due che stiamo per scrivere serviranno per determinare numericamente la media e la varianza nei due 
# casi di uniform e jeffrey's prior

uni.post.fun <- function(x,data=deaths) dgamma(x, shape=uniform.posterior(data)[1], rate=uniform.posterior(data)[2])
jef.post.fun <- function(x,data=deaths) dgamma(x, shape=jeffrey.posterior(data)[1], rate=jeffrey.posterior(data)[2])

uni.post.fun.mean <- function(x) return(x*uni.post.fun(x))
uni.post.fun.var <- function(x) return((x^2)*uni.post.fun(x))
jef.post.fun.mean <- function(x) return(x*jef.post.fun(x))
jef.post.fun.var <- function(x) return((x^2)*jef.post.fun(x))
```

```r
alpha.uniform <- uniform.posterior(deaths)[1]
lambda.uniform <- uniform.posterior(deaths)[2]
alpha.jeffrey <- jeffrey.posterior(deaths)[1]
lambda.jeffrey <- jeffrey.posterior(deaths)[2]

gamma.mean.uniform <- alpha.uniform/lambda.uniform
gamma.variance.uniform <- alpha.uniform/(lambda.uniform^2)
gamma.mean.jeffrey <- alpha.jeffrey/lambda.jeffrey
gamma.variance.jeffrey <- alpha.jeffrey/(lambda.jeffrey^2)
gamma.mode.uniform <- compute.mode.gamma(alpha.uniform,lambda.uniform)
gamma.mode.jeffrey <- compute.mode.gamma(alpha.jeffrey,lambda.jeffrey)
```

```r
sprintf("Prior uniforme")
sprintf("Calcoli analitici: media=%.4f, varianza=%.4f, moda=%.4f", gamma.mean.uniform, gamma.variance.uniform, gamma.mode.uniform)

sprintf("Prior di Jeffrey")
sprintf("Calcoli analitici: media=%.4f, varianza=%.4f, moda=%.4f", gamma.mean.jeffrey, gamma.variance.jeffrey, gamma.mode.jeffrey)
```

### Intervallo di credibilità 
Analogamente a quanto visto nell'esercizio 5 stimiamo inizialmente l'intervallo di credibilità del 95% graficamente per entrambi i prior e successivamente confrontiamo il risultato con l'approssimazione gaussiana.

$$\int_{x_1}^{x_2}P(\mu|\{x_j\}) = 95\% \ \ \ \ \text{contro l'approssimazione} \\ P(\mu|\{x_j\}) \approx A \exp \left[\frac{1}{2} \frac{\mathbf{d^2}L}{\mathbf{d}\mu^2}\Bigr|_{\substack{\mu = \mu_0}} (\mu - \mu_0)^2 \right] \\ \text{con} \ \ L=\log(P(\mu|\{x_j\}))$$

Da questa relazione deduciamo che la devizione standard della gaussiana che meglio approssima la distribuzione è pari a 

$$ \sigma = \left(-\frac{1}{2} \frac{\mathbf{d^2}L}{\mathbf{d}\mu^2}\Bigr|_{\substack{\mu = \mu_0}}\right)^{-1/2}$$

pertanto la probabilità che il parametro $\mu_0$ giaccia all'interno di un intervallo $\mu_0 \pm 2 \sigma$ sarà del 95% secondo la misura di un integrale gaussiano entro due devizioni standard dal valor medio della distribuzione.

I valori di $\mu$ e $\sigma$ sono scelti uguali a quelli delle distribuzioni Gamma restituite come posterior dall'inferenza baiesiana.

```r
interval.plot.line <- function(a,b,col){
    if(is.character(col)){
        abline(v=a, col=col, lwd=3)
        abline(v=b, col=col, lwd=3)
        }
    else cat("colore errato")
}
```

```r
# vediamo quanto valgono gli estremi degli intervalli di credibilità per i posterior uniforme e di jeffrey 
a.uniform <- gamma.mode.uniform - 2*sqrt(gamma.variance.uniform); b.uniform <- gamma.mode.uniform + 2*sqrt(gamma.variance.uniform)
a.jeffrey <- gamma.mode.jeffrey - 2*sqrt(gamma.variance.jeffrey); b.jeffrey <- gamma.mode.jeffrey + 2*sqrt(gamma.variance.jeffrey)
```

```r
plot(x,dgamma(x, shape=alpha.uniform, rate=lambda.uniform), type = "l", lwd = 3, col = "dodgerblue", xlab=expression(mu),ylab="P(mu|x)", 
     main="Posterior", xlim = c(0.5,1.5))
lines(x,dgamma(x, shape=alpha.jeffrey, rate=lambda.jeffrey), type = "l", lwd = 3, col = "blue")
legend("topright", legend=c("Gamma posterior da Jeffrey's prior","Gamma posterior da prior uniforme", "Int credibilità Jeffrey",
                           "Int credibilità Uniform"), 
       col=c("blue", "dodgerblue", "orange", "red"),
      fill = c("blue", "dodgerblue", "orange", "red"))

interval.plot.line(a.uniform,b.uniform,"red")
interval.plot.line(a.jeffrey,b.jeffrey,"orange")

sprintf("I due intervalli di credibilità estratti graficamente corrispondono al %.3f %% e %.3f %%", integrate(uni.post.fun, a.uniform,b.uniform)$value*100, 
        integrate(jef.post.fun, a.jeffrey,b.jeffrey)$value*100)
```

## Esercizio 3

Per capire quanti canali hanno un alto livello di batteri $X$, sono stati presi $n_0=116$ campioni. Tra questi soltanto $y_0=11$ presentano un alto livello di batteri $X$, pertanto la stima frequentista del parametro $p_0$, ovvero la probabilità di osservare un elevata quantità di batteri in un campione, equivale a $p_0=y_0/n_0$ 

```r
# numero di campioni considerati
n0 <- 116

# tra n0 numero di campioni con alto livello di batteri X
y0 <- 11
p0 <- y0/n0
sprintf("Stima frequentista, p=%.2f",p0)
```

#### Beta(1,10) prior per p al fine di calcolare la posterior distribution

Ancora una volta il teorema di Bayes può essere applicato usando come prior una distribuzione $\mathrm{Beta}(\alpha_{prior},\beta_{prior})$ con $\alpha_{prior}=1$ e $\beta_{prior}=10$. La funzione posterior può essere ricavata sfruttando la proprietà dei conjugate priors ovvero avendo scelto come prior una distribuzione beta di parametri $\alpha,\beta$ allora il posterior coinciderà con un'altra $\mathrm{Beta}$ distribution i cui parametri saranno aggiornati grazie ai dati raccolti:

$$ \alpha_{posterior}=\alpha_{prior} + y_0 = 12 \\
\beta_{posterior}=\beta_{prior} + n_0 - y_0 = 115$$


```r
alpha_prior0 <- 1
beta_prior0 <- 10
prior0 <- function(t) dbeta(t,shape1=alpha_prior0,shape2=beta_prior0)

alpha_posterior0 <- alpha_prior0+y
beta_posterior0 <- beta_prior0+n-y
posterior0 <- Vectorize(function(t){dbeta(t,shape1=alpha_posterior0,shape2=beta_posterior0)})

p <- seq(0,0.25,0.001)

plot(p,posterior0(p),type="l",col="darkgreen",main="Distribuzioni di probabilità",ylab="Densità di probabilità",lwd=3)
lines(p,prior0(p),col="darkblue",type="l",lwd=3)
legend("topright",legend=c("Posterior","Prior"),col=c("darkgreen","darkblue"),fill=c("darkgreen","darkblue"))
```

#### Trovare la stima bayesiana per p, la moda, varianza e un intervallo di credibilità del 95%



```r
mean <- integrate(function(t){t*posterior0(t)},lower=0,upper=1)$value

variance <- integrate(function(t){posterior0(t)*(t-mean)^2},lower=0,upper=1)$value

norm_posterior <- Vectorize(function(p){dnorm(p,mean=mean,sd=sqrt(variance))})

upper_bayes <- qnorm(0.975,mean=mean,sd=sqrt(variance))

lower_bayes <- qnorm(0.025,mean=mean,sd=sqrt(variance))

credibility <- integrate(function(t){dnorm(t,mean=mean,sd=sqrt(variance))},lower=lower_bayes,upper=upper_bayes)$value

p <- seq(0,0.25,0.001)
plot(p,norm_posterior(p),type="l",col="red",main="Approssimazione gaussiana del posterior",lwd=3,ylab="Densità di probabilità")
legend("topright",legend=sprintf("Intervallo di credibilità del 95%%"), 
      col="blue",fill="blue")
polygon(x=c(lower_bayes,lower_bayes,p[(p>=lower_bayes)&(p<=upper_bayes)],upper_bayes),y=c(0,norm_posterior(lower_bayes),norm_posterior(p[(p>=lower_bayes)&(p<=upper_bayes)]),0),col="blue")
abline(v=c(lower_bayes,mean,upper_bayes),col="black")
                            
```

#### Hypothesis test $H_0: p=0.1$ contro $H_1: p\neq 0.1$

- L'ipotesi nulla $H_0$ descrive che il parametro $p$ è uguale a 0.1
- $H_1$ invece afferma che $p\neq 0.1$

**Approccio bayesiano** 

Per eseguire l'hypothesis test è necessario innanzitutto individuare una regione di "acceptance" nella distribuzione del posterior, in questo caso utilizzo direttamente i boundaries trovati attraverso l'approssimazione gaussiana di un intervallo di credibilità del 95%. Poichè $p=0.1$ giace dentro questo intervallo allora non è possibile rigettare $H_0$.

**Approccio frequentista**

Seguendo invece il paradigma frequentista rappresentiamo la verosimiglianza binomiale nelle condizioni in cui $H_0$ sia vera, troviamo i due estremi dell'intervallo di "accettanza" in modo da individuare la regione di rigetto della null hypothesis con un p-value il più vicino possibile a $\alpha=0.05$

- Intervallo di rigetto di $H_0$: ${k<k_{min}=6} \cup {k>k_{max}=20}$
- Probabilità totale: $\alpha=0.044$

poichè $y_0=11$ giace in questa regione non possiamo escludere $H_0$. 

```r
#BAYES
if((0.1<=upper_bayes)&(0.1>=lower_bayes)){sprintf("BAYES: non possiamo rigettare H0")}

p <- seq(0,0.25,0.001)
plot(p,posterior0(p),type="l",col="darkred",main="BAYESIAN: Posterior Distribution con la regione di rigetto",lwd=3,ylab="Densità di probabilità")
polygon(x=c(upper_bayes,upper_bayes,p[p>=upper_bayes],0.25),y=c(0,posterior0(upper_bayes),posterior0(p[p>=upper_bayes]),0),col="red")
polygon(x=c(0,p[p<=lower_bayes],lower_bayes,lower_bayes),y=c(0,posterior0(p[p<=lower_bayes]),posterior0(lower_bayes),0),col="red")
abline(v=c(lower_bayes,upper_bayes),col="black")
abline(v=0.1,col="red")
legend("topright",legend=c("Estremi di rigetto","p=0.1"),col=c("black","red"),lty=1:1)
                                        
    
#FREQUENTIST

H0 <- function(t) dbinom(t,size=n0,prob=0.1)
pvalue_low <- function(t) sum(dbinom(x=seq(0,n,1),size=n,prob=p0)[1:t])
k_min <- 6
lower <- pvalue_low(k_min)
pvalue_up <- function(t) sum(dbinom(x=seq(0,n,1),size=n,prob=p0)[t:n])
k_max <- n0-96
upper <- pvalue_up(k_max)
alpha_dist <- lower+upper

if((y<k_max)&(y>=k_min)){sprintf("FREQUENTIST: non possiamo rigettare H0")}

#plot
x <- seq(0,n,1)
plot(x,H0(x),type="l",col="darkred",lwd=3,main="FREQUENTIST: Binomial Distribution (H0=True) con regione di rigetto",ylab="Densità di probabilità",xlab="k")
polygon(x=c(0,x[x<=k_min],k_min,k_min),y=c(0,H0(x[x<=k_min]),H0(k_min),0),col="red")
polygon(x=c(k_max,k_max,x[x>=k_max],n),y=c(0,H0(k_max),H0(x[x>=k_max]),0),col="red")
abline(v=c(6,n0-96),col="black")
abline(v=y0,col="red")
legend("topright",legend=c("Estremi di rigetto","k=11"),col=c("black","red"),lty=1:1)

```

#### Stima frequentista per p

Date le nuove informazioni ($y_1=9$) ottenute misurando altri samples, $n1=165$, possiamo decidere se aggiornare l'informazione che avevamo finora archiviato oppure trattare queste misure come un set ex novo.

Se decidessimo di aggiornare i parametri che abbiamo calcolato fino a questo momento, usando pertanto i due set di misure insieme, la nuova stima frequentista per il parametro $p$ sarebbe $p=\frac{1}{2}\left(\frac{y_0}{n_0}+\frac{y_1}{n_1}\right) = 0.07$. Considerando soltanto il nuovo set di misure invece l'estimator sarebbe $p=\frac{y_1}{n_1}=0.05$

```r
n1 <- 165
y1 <- 9

p1 <- (y0/n0+y1/n1)/2

sprintf("Stima del parametro p aggiornando le precedenti informazioni = %.2f",p1)
sprintf("Stima del parametro p usando il nuovo set come unico set di misure = %.2f",y1/n1)
```

#### Stima di p con approccio baiesiano in due condizioni: usando un prior $\mathrm{Beta}(1,10)$ oppure utilizzando il precedente posterior come prior per il nuovo processo di inferenza

```r
#BETA PRIOR

alpha_prior1 <- 1
beta_prior1 <- 10
prior1 <- Vectorize(function(t){dbeta(t,shape1=alpha_prior1,shape2=beta_prior1)})
alpha_posterior1 <- alpha_prior1+y1
beta_posterior1 <- beta_prior1+n1-y1
posterior1 <- Vectorize(function(t){dbeta(t,shape1=alpha_posterior1,shape2=beta_posterior1)})
p <- seq(0,0.25,0.001)
plot(p,posterior1(p),type="l",col="dodgerblue",main="Posterior con beta prior",ylab="Densità di probabilità",lwd=3)
lines(p,prior1(p),col="blue",type="l",lwd=3)
legend("topright",legend=c("Posterior","Prior"),col=c("dodgerblue","blue"),lty=1:1)


#UPDATED PRIOR

prior1 <- posterior0
alpha_posterior1 <- alpha_posterior0+y1
beta_posterior1 <- beta_posterior0+n1-y1
posterior1 <- Vectorize(function(t){dbeta(t,shape1=alpha_posterior1,shape2=beta_posterior1)})
p <- seq(0,0.25,0.001)
plot(p,posterior1(p),type="l",col="dodgerblue",main="Posterior con prior=precedente posterior",ylab="Densità di probabilità",lwd=3)
lines(p,prior1(p),col="blue",type="l",lwd=3)
legend("topright",legend=c("Posterior","Prior"),col=c("dodgerblue","blue"),lty=1:1)
```

#### Trovare una stima baiesiana per p, moda, varianza e un intervallo di credibilità al 95%

```r
# da questo momento in poi utilizzerò soltanto il posterior aggiornato tramite il prior uguale al posterior del precedente processo di
# inferenza 

mean <- integrate(function(t){t*posterior1(t)},lower=0,upper=1)$value
variance <- integrate(function(t){posterior1(t)*(t-mean)^2},lower=0,upper=1)$value
norm_posterior <- Vectorize(function(p){dnorm(p,mean=mean,sd=sqrt(variance))})
upper_bayes <- qnorm(0.975,mean=mean,sd=sqrt(variance))
lower_bayes <- qnorm(0.025,mean=mean,sd=sqrt(variance))
credibility <- integrate(function(t){dnorm(t,mean=mean,sd=sqrt(variance))},lower=lower_bayes,upper=upper_bayes)$value
p <- seq(0,0.25,0.001)
plot(p,norm_posterior(p),type="l",col="dodgerblue",main="Approssimazione gaussiana del posterior",lwd=3,ylab="Densità di probabilità")
legend("topright",legend="Intervallo di credibilità 95%%",col="green",fill="green")
polygon(x=c(lower_bayes,lower_bayes,p[(p>=lower_bayes)&(p<=upper_bayes)],upper_bayes),y=c(0,norm_posterior(lower_bayes),norm_posterior(p[(p>=lower_bayes)&(p<=upper_bayes)]),0),col="green")
abline(v=c(lower_bayes,mean,upper_bayes),col="black")
```
